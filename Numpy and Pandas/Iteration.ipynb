```python
df2 = pd.DataFrame(students_dict, index = [2, 3, 5, 6, 8, 9, 11, 12, 14, 15])
```

```python
df = df.append(df2, ignore_index=True)
```

```python
df
```

```python
dff = df.iloc[0:5, 0:3]
```

```python
dff
```

## Iterating over rows

```python
for row_label, row in dff.iterrows():
    print(row_label, row)
```

```python
for row_label, row in dff.iterrows():
    print(row_label, row, sep='\n', end='\n\n') # (Index, Series)
```

```python
start = time.time()

for row_label, row in dff.iterrows():
    print(row_label, row, sep='\n', end='\n\n') # (Index, Series)
   
print(f'Total time is {time.time() - start}')
```

```python
start = time.time()

for row in dff.itertuples():
    print(row, sep='\n', end='\n') # (Index, Series)
   
print(f'Total time is {time.time() - start}')

# itertuples are wayy faster than iterrows.
```

```python
start = time.time()

for row_label, row in df.iterrows():
    print(row_label, row, sep='\n', end='\n\n') # (Index, Series)
   
print(f'Total time is {time.time() - start}')
```

```python
start = time.time()

for row in df.itertuples():
    print(row, sep='\n', end='\n') # (Index, Series)
   
print(f'Total time is {time.time() - start}')
```

## Combining Data on a Common Column Index

```python
detail1 = {'key1': ['k0', 'k1', 'k2', 'k3'],
           'key2': ['k0', 'k1', 'k0', 'k1'],
           'name': ['Ibrahim Musa', 'Emeka Jude', 'Emmanuel Ayomide', 'Nnamdi Ike'],
           'age': [24, 25, 27, 24]}
```

```python
detail2 = {'key1': ['k0', 'k1', 'k2', 'k3'],
           'key2': ['k0', 'k0', 'k0', 'k0'],
           'Course': ['Physics', 'Statistics', 'Mathematics', 'Chemistry'],
           'Sport': ['Football', 'Volleyball', 'Swimming', 'Basketball']}
```

```python
dff1 = pd.DataFrame(detail1)
```

```python
dff2 = pd.DataFrame(detail2)
```

```python
dff1
```

```python
dff2
```

```python
pd.merge(dff1, dff2, on=['key1'])
```

```python
pd.merge(dff1, dff2, on=['key1', 'key2'])
```

```python
pd.merge(dff1, dff2, how='inner', on=['key1', 'key2'])
```

```python
pd.merge(dff1, dff2, how='left', on=['key1', 'key2'])
```

```python
pd.merge(dff1, dff2, how='right', on=['key1', 'key2'])
```

```python
pd.merge(dff1, dff2, how='outer', on=['key1', 'key2'])
```
